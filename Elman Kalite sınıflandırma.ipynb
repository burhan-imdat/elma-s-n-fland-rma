```python
from pycaret.classification import*
# gerekli kütüphane indirildi
```

```python
df=pd.read_csv('elma.csv')
# dosya okundu
```

```python
df.head()
# dosya içerik il 5 satır gösterildi
```

```python
df=df.drop('A_id',axis=1)
# gerekli görülmeyen sutun silindi 
```

```python
df.isnull().sum()
# boş değer varlığı kontrol edildi
```

```python
df.shape
# sutun ve satır sayısına bakıldı
```

```python
df.dropna(inplace=True)
# boş değerler mevcut listeden düşürüldü  inplace=yapılan değişiklikler kayıt edilir
```

```python
setup(data=df,target='Quality')
# hedef geğikene göre makine öğrenme sistemi kuruldu  (sınıflandırma)
```

```python
compare_models()
# sınıflandırma modeller tatbik edildi (denendi)
```

```python
best_model=create_model('knn')
# en iyi treen test kombinasyonuu bulundu
```

```python
import joblib  
# model kayıt için kütüphane indirildi
```

```python
joblib.dump(best_model,'elmakalite.joblib')
# model kayıt edildi
```
